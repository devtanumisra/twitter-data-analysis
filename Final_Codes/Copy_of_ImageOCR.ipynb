```python
import pytesseract
from PIL import Image
```

```python
f = open('twitter_data_ocr.csv', 'w', encoding='UTF8',newline='')
writer = csv.writer(f)
header = ['POST_ID','IMAGE_ID','Text']
writer.writerow(header)
```

```python
from zmq.constants import NULL
for image in images.iterrows():
  print(image[1]['POST_ID'])
  print(image[1]['IMAGES_ID'])
  file1 = folder_dir+str(image[1]['IMAGES_ID'])

  img=Image.open(file1)

  length_x, width_y = img.size
  factor = min(1, float(1024.0 / length_x))
  size = int(factor * length_x), int(factor * width_y)
  im_resized = img.resize(size, Image.ANTIALIAS)
  text=pytesseract.image_to_string(im_resized,lang='eng')

  post_row = [image[1]['POST_ID'], image[1]['IMAGES_ID'], text if text!='\n' else "NA"]
  writer.writerow(post_row)
  print(text)
  print()
```
