```python
from transformers import pipeline
```

```python
caption = pipeline('image-to-text')
```

```python
import os
import csv
from os import listdir
from IPython.display import Image

list1=[]
f = open('twitter_data_img_penaldo.csv', 'w', encoding='UTF8',newline='')
writer = csv.writer(f)
header = ['POST_MEDIA','POST_CAPTION']
writer.writerow(header)
# get the path or directory
folder_dir = "/content/drive/MyDrive/Penaldo/"
i = 0
for images in os.listdir(folder_dir):

    # check if the image ends with png or jpg or jpeg
    if (images.endswith(".png") or images.endswith(".jpg")\
        or images.endswith(".jpeg")):
        # display
        file1 = folder_dir+images
        output = caption(file1)
        post_row = [images, output]
        writer.writerow(post_row)
        print(output)
        i=i+1
        if(i == 1):
          break

f.close()
```
