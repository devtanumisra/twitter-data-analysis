## OPEN A CSV FILE AND WRITE THE RESPONSE TO IT

```python
import csv
```

```python
header = ['PID', 'Text', 'Media', 'MediaURL', 'Emotion', 'HateSpeech']
```

```python
for trump in trumps:
    st = trump.text
    index = st.rfind("https")
    print("NA" if (index == -1) else st[index:])
```

```python
with open('twitter_data_2.csv', 'w', encoding='UTF8',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)

    # write the data
    for trump in trumps:
        st = trump.text
        print(type(st))
        index = st.rfind("https")
        print(index)
        print(st[index:])
        row = ["P:"+str(trump.id), trump.text, "NA" if (index == -1) else st[index:]]
        writer.writerow(row)
```
